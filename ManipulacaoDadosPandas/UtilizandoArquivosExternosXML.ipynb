#### Utilizando arquivos externos para tratamento de dados

```python
import pandas as pd

Path = "../MaterialEstudos/BaseDeDados/"
File = "bank-full.xlsx"

#Carregando o arquivo para o DataFrame
df = pd.read_excel(Path+File, sheet_name="Sheet 1 - bank-full")

print(df) # Quebra automaticamente a linha, resestruturando a tabela na parte inferior
df # Apresenta a tabela de dados sem quebra de linha
```

#### Trabalhando com as partes do dataframe, definindo a quantidade de linha que se deseja exibir (Quantidade passa como parametro)

```python
df.head(2) # retorna a parte superior da tabela com uma linha de código
```

```python
df.tail(1) # retorna a parte inferior da tabela com uma linha de código
```

```python
df.info() # retorna as infornações de todas as colunas, como NÚMERO DE REGISTROS NA COLUNA, SE É NOT NULL, E QUAL O TIPO DE DADOS TRABALHADOS NA COLUNA
```

```python
df.describe() # Retorna estatísticas relacionados ao resumo por linhas, como: COUNT, MEAN, STD e Outros
```

```python
df.describe(include=["O"]) # Retorna estatísticas relacionados a texto, como: NÚMERO DE VEZES QUE APARECE, QUANTOS VALORES ÚNICOS EXISTEM, QUAIS AS PALAVRAS MAIS REPETIDAS
#count - Contagem | unique - quantidade de elementos únicos
#top - palavra que aparece com maior frequencia
#freq - quantidade de vezes que a palavra aparece
```

```python
df.describe(include="all") # retorna dados estatísticos de categoricos(texto) e numéricos
```
