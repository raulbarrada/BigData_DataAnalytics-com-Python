```python
#Importanto bibliotecas Numpay e MatPlot
import numpy as np
import matplotlib.pyplot as plt
#Criando array que passará os valores de X e Y
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
#Criando um gráfico de linha simples, mas modificando o tipo de marcador
plt.plot(x,y,marker = "*", c = "#fd28af") #Onde marker(valores aceitos *, o, x) define o cruzamento dos eixos e C a cor da linha (cor em ingles ou hexadecimal)
```

```python
plt.plot(
    x, y,               # Definindo eixos X e Y
    marker = "o",       # Formato dos cruzamentos dos eixos
    c = "#f00",         # Define a cor da linha
    ds = "steps",       # Estilo da ligação dos pontos
    ls = "--",          # Estilo da linha
    lw=3,               # Espessura da linha
    mec = "#f00",       # Define a cor da borda dos pontos
    mfc = "#000"        # Define a cor interna do marcador
    )

#Adicionando título para o eixo X e Y
plt.xlabel("Título X", fontsize = 16)
plt.ylabel("Título Y", fontsize = 16)

#Adicionando título do gráfico
plt.title("Título do gráfico", fontsize=20, c = "#333")

#Alterando os rótulos do eixos
plt.xticks(fontsize=8, c = "purple")
plt.yticks(fontsize=8, c = "orange")
```

```python
# TRABALHANDO GRÁFICOS COM MULTIPLAS CURVAS E LEGENDAS

x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label = "SENO")                      # Label indica o nome das curvas e será exibido na legenda do gráfico
plt.plot(x, z, label = "COSSENO")
plt.plot(x, np.sqrt(x), label = "RAÍZ QUADRADA")

plt.legend()                                        #Adiciona a legenda no gráfico
```

```python
# TRABALHANDO SUB GRÁFICOS
# Configurando o primeiro subgráfico

plt.subplot(3,2,1) # (quantidade horizontal, quantidade vertical, gráfico a ser trabalhado)
plt.title("Sub Gráficos 01")
plt.plot(x, y)


# Configurando o segundo subgráfico
plt.subplot(3,2,2) # (quantidade horizontal, quantidade vertical, gráfico a ser trabalhado)
plt.title("Sub Gráficos 02")
plt.plot(x, z)

# Configurando o terceiro subgráfico

plt.subplot(3,2,3) # (quantidade horizontal, quantidade vertical, gráfico a ser trabalhado)
plt.title("Sub Gráficos com multiplas linhas")
plt.plot(y, z)
plt.plot(x, z)
```
