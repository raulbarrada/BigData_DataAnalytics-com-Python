```python
import pandas as pd
import matplotlib.pyplot as plt
import os # Gerencia o sistema operacional

PathFile = "../MaterialEstudos/BaseDeDados/"
NameFile = "company_sales_data.csv"

Datas = pd.read_csv(PathFile+NameFile)

Datas.head(0)
```

```python
#Criando uma listas para cada coluna
ListMonths      = Datas["month_number"].to_list()
ListProfit      = Datas["total_profit"].to_list()
ListUnits       = Datas["total_units"].to_list()
ListFaceCream   = Datas["facecream"].to_list()
ListFaceWash    = Datas["facewash"].to_list()
ListToothPaste  = Datas["toothpaste"].to_list()
ListBathing     = Datas["bathingsoap"].to_list()
ListShampoo     = Datas["shampoo"].to_list()
ListMoisturizer = Datas["moisturizer"].to_list()
```

```python

plt.plot(
    ListMonths, 
    ListFaceCream, 
    label = "Creme facial",
    c = "#a5f",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0"
    )
plt.plot(
    ListMonths,             #Eixo X
    ListFaceWash,             #Eixo Y
    label = "Limpador facial",        #Rótulo do gráfico
    c = "#adc",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0"
    )
plt.plot(
    ListMonths, 
    ListBathing, 
    label = "Material Banho",
    c = "#dca",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0",
    )
plt.plot(
    ListMonths,             #Eixo X
    ListToothPaste,           #Eixo Y
    label = "Creme dental",        #Rótulo do gráfico
    c = "#a81",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0"
    )

plt.plot(
    ListMonths, 
    ListBathing, 
    label = "Material Banho",
    c = "#123",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0",
    )
plt.plot(
    ListMonths, 
    ListMoisturizer, 
    label = "Hidratante",
    c = "#32a",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0",
    )

plt.plot(
    ListMonths, 
    ListShampoo, 
    label = "Shampoo",
    c = "#8a1",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0",
    )



plt.title("Análise de Lucro Mensal (LucroProduto/Mês)", fontsize=14, fontweight="bold", c = "#af3d9c")
plt.xlabel("Meses")
plt.ylabel("Lucro Total")

plt.legend(loc='upper left', fontsize = 8)

os.mkdir("./GraphGenerateTask02")
plt.savefig("./GraphGenerateTask02/gráfico")
```
