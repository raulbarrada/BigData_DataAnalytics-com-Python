```python
import pandas as pd
import matplotlib.pyplot as plt
import os # Gerencia o sistema operacional

PathFile = "../MaterialEstudos/BaseDeDados/"
NameFile = "company_sales_data.csv"

Datas = pd.read_csv(PathFile+NameFile)

Datas.head(0)

```

```python
#Criando uma listas para cada coluna
ListMonths = Datas["month_number"].to_list()
ListUnits = Datas["total_units"].to_list()
ListFaceCream = Datas["facecream"].to_list()
ListFaceWash = Datas["facewash"].to_list()
ListToothPaste = Datas["toothpaste"].to_list()

#Criando o gráfico
plt.plot(ListMonths, ListUnits, ListFaceCream)
#plt.plot(ListMonths, ListFaceCream)
#plt.plot(ListMonths, ListFaceWash)
#plt.plot(ListMonths, ListToothPaste)
```

```python
#Criando uma listas com o lucro total de todos os meses
ListMonths = Datas["month_number"].to_list()
ListProfit = Datas["total_profit"].to_list()

#Criando o gráfico personalizado
plt.plot(
    ListMonths,             #Eixo X
    ListProfit,             #Eixo Y
    label = "Dados do lucro dos últimos meses",        #Rótulo do gráfico
    c = "#ab244f",          #Cor da linha do gráfico
    ls = "--",              #Estilo da linha
    lw = 3,                 #Espessura da linha
    marker = "o",           #Tipo do marcador
    mfc = "#0f0",
    )

plt.title("Análise de Lucro Mensal (Lucro/Mês)", fontsize=14, fontweight="bold", c = "#af3d9c")
plt.xlabel("Meses")
plt.ylabel("Lucro Total")

plt.legend()

#os.mkdir("./GraphGenerateTask02")
plt.savefig("./GraphGenerateTask02/gráfico")
```
