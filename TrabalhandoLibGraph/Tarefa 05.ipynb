```python
import pandas as pd
import matplotlib.pyplot as plt
import os # Gerencia o sistema operacional

PathFile = "../MaterialEstudos/BaseDeDados/"
NameFile = "company_sales_data.csv"

Datas = pd.read_csv(PathFile+NameFile)

Datas.head(0)
```

```python
#Criando uma listas para cada coluna
ListMonths      = Datas["month_number"].to_list()
ListFaceCream   = Datas["facecream"].to_list()
ListToothPaste  = Datas["toothpaste"].to_list()
```

```python
#Criando gráfico de dispersão
plt.bar(
    [pos-0.25 for pos in ListMonths],   #DESLOCAR 0.25 À ESQUERDA
    ListToothPaste,                     #VALOR DE Y
    label = "Creme dental",
    width = 0.25,
)
plt.bar(
    [pos+0.25 for pos in ListMonths], #DESLOCAR 0.25 À DIREITA
    ListFaceCream,           #VALOR DE Y
    label = "Creme facial",
    width = 0.25,    
)

#Adicionando títulos no gráfico
plt.title("Gráfico de Barra - Mais vendidos", fontsize=14, fontweight="bold", c = "#c56")
plt.xlabel("Meses")
plt.ylabel("Unidades vendidas")

os.mkdir("./GraphGenerateTask05")
plt.savefig("./GraphGenerateTask05/gráfico")
```
